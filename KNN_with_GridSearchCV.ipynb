```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.datasets import load_wine

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score

from sklearn.metrics import confusion_matrix,classification_report

from sklearn.model_selection import GridSearchCV
```

```python
df=load_wine(as_frame=True).frame

x = df.drop(['target'],axis=1)
y = df['target']

x_train , x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
knn_model= Pipeline(
    steps=[
        ('scaler',StandardScaler()),
        ('knn',KNeighborsClassifier())

    ]
)

knn_model.fit(x_train,y_train)
```

```python
# cross validation er maddhome best parameter ber korar jonno gridsearchcv use kora hoy
parameter_grid = {
    'knn__n_neighbors': range(1,30),
    'knn__weights': ['uniform', 'distance'],
    'knn__p': [1, 2]
}

grid_search = GridSearchCV(
    estimator=knn_model,
    param_grid=parameter_grid,
    cv=5,
)


# param_griddict or list of dictionaries. ...ata dict format tai baire akta dictionary create kore kaj korsi

# sklear gridserchcv dile estimator er pr konta kibabe dia eta jante prba google e

grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_
```

```python
model=grid_search.best_estimator_

y_pred=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)

print(f"Accuracy: {accuracy}")
```
